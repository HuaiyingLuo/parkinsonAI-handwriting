```python
# calculate the size of an image
dir = "../resnet50/handwritten_dataset/meander/HealthyMeander"

import os
from PIL import Image
import numpy as np

for file in os.listdir(dir):
    img = Image.open(os.path.join(dir, file))
    print(img.size)

# get the average size of the images
sizes = []
for file in os.listdir(dir):
    img = Image.open(os.path.join(dir, file))
    sizes.append(img.size)

print(np.mean(sizes, axis=0))

# this shows the size is not consistent
```
